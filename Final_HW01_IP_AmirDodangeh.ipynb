```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

figsize = (10,10)


```

First, read the image as grayscale



```python






# read as grayscale
I = cv2.imread("image3.jpg",0)

plt.figure(figsize=figsize)
plt.imshow(image3  , cmap='gray' , vmin=0 , vmax=255)
plt.title("Original image")
```

Let's start by calculating and showing the original histogram


```python
bins_edges_min_max = [0,256]
num_bins=256
bin_count,bins_edges = np.histogram(image3,num_bins,bins_edges_min_max)
bins_start = bins_edges[:-1]
```

```python
def draw_hist(x_axis,input):
    fig,ax = plt.subplots(figsize=figsize)
    # why not using plt.hist? because we want to plot also some derivations of this hist, so this is easier
    plt.bar(x_axis, input, width=input.shape[0]/(x_axis[-1]-x_axis[0]+1))
    return fig,ax

draw_hist(bins_start,bin_count)
plt.title("Original histogram")

```

Normalize the histogram to gat a discrete PDF


```python
pdf = bin_count/np.sum(bin_count)

draw_hist(bins_start,pdf)
plt.title("Original PDF")

```

Get the CDF by calculating the cumulative sum of the pdf data


```python
cdf = np.cumsum(pdf)

plt.figure(figsize=figsize)
plt.plot(cdf)
plt.title("Original CDF")

```

```python
fig,ax = draw_hist(bins_start,pdf)
ax.plot(cdf*np.max(pdf),'r')
plt.title("Original PDF+ const*CDF to show the connection between the two")

```

The final step is to un-normalize the CDF to become the equalization function


```python
f_eq = np.round(cdf*255).astype(int)

f_eq
```

Use the equalization function to get the equalized image


```python
I_eq = f_eq[image3]

plt.figure(figsize=figsize)
plt.imshow(I_eq, cmap='gray', vmin=0, vmax=255)
plt.title("equalized image")
```

Plot the equalized histogram, PDF and CDF


```python
bin_count,bins_edges = np.histogram(I_eq,num_bins,bins_edges_min_max)
bins_start = bins_edges[:-1]

draw_hist(bins_start,bin_count)
plt.title("equalized histogram")
```

```python
pdf = bin_count/np.sum(bin_count)
cdf = np.cumsum(pdf)

fig,ax = draw_hist(bins_start,pdf)
ax.plot(cdf*np.max(pdf),'r')
plt.title("equalized PDF and const*CDF")
```

```python
import numpy as np
%matplotlib inline
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import data
from skimage import exposure
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
from skimage.exposure import match_histograms

#image3=io.imread("/content/gdrive/My Drive/Colab Notebooks/Image Processing/HW1/image3.jpg",0)

source=image3
#source=cv.cvtColor(image3, cv.COLOR_BGR2GRAY)

reference=I_eq_cv2


#reference=cv.cvtColor(reference, cv.COLOR_BGR2GRAY)


#cv2_imshow(source)
#cv2_imshow(reference)

matched = match_histograms(source, reference, multichannel=True)

cv2_imshow(source)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(source , cmap='gray', vmin=0, vmax=255)
ax1.set_title('Source')
ax2.imshow(reference , cmap='gray', vmin=0, vmax=255)

ax2.set_title('Reference')
ax3.imshow(matched , cmap='gray', vmin=0, vmax=255)
ax3.set_title('Matched')

plt.tight_layout()
plt.show()
```

```python
bin_count,bins_edges = np.histogram(source,num_bins,bins_edges_min_max)
bins_start = bins_edges[:-1]

draw_hist(bins_start,bin_count)
plt.title("Source")

bin_count,bins_edges = np.histogram(reference,num_bins,bins_edges_min_max)
bins_start = bins_edges[:-1]

draw_hist(bins_start,bin_count)
plt.title("Reference")


bin_count,bins_edges = np.histogram(matched,num_bins,bins_edges_min_max)
bins_start = bins_edges[:-1]

draw_hist(bins_start,bin_count)
plt.title("Matched")
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(8, 8))


for i, img in enumerate((source, reference, matched)):
    for c, c_color in enumerate(('red', 'green', 'blue')):
        img_hist, bins = exposure.histogram(img[..., c], source_range='dtype')
        axes[c, i].plot(bins, img_hist / img_hist.max())
        img_cdf, bins = exposure.cumulative_distribution(img[..., c])
        axes[c, i].plot(bins, img_cdf)
        axes[c, 0].set_ylabel(c_color)

axes[0, 0].set_title('Source')
axes[0, 1].set_title('Reference')
axes[0, 2].set_title('Matched')

plt.tight_layout()
plt.show()
```
