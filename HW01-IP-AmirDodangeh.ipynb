Histogram equalization:
1- First, display Gaussian Probability Mass Function (PMF) and Cumulative Mass Function (CMF)
with mu=0 and sigma=1 second, apply histogram equalization and histogram matching transform in
image3 third, plot each histogram and compare them.


A probability mass function differs from a probability density function (PDF) in that the latter is associated with continuous rather than discrete random variables. A PDF must be integrated over an interval to yield a probability. The value of the random variable having the largest probability mass is called the mode

             ({\displaystyle \mu }\mu )      ({\displaystyle \sigma }\sigma )  .         .     {\displaystyle \mu =0}{\displaystyle \mu =0}  {\displaystyle \sigma =1}{\displaystyle \sigma =1}      .

```python
import numpy as np
%matplotlib inline
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt

image3=io.imread("/content/gdrive/My Drive/Colab Notebooks/Image Processing/HW1/image3.jpg",0)
image3=cv.cvtColor(image3, cv.COLOR_BGR2GRAY)
cv2_imshow(image3)
print(image3.shape)
print(image3.dtype)

plt.hist(image3.ravel(),256,[0,256]); plt.show()


eq_grayscale_image = cv.equalizeHist(image3)
cv2_imshow(eq_grayscale_image)

plt.hist(eq_grayscale_image.ravel(),256,[0,256]); plt.show()





hist,bins = np.histogram(image3.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(image3.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()


cdf_m = np.ma.masked_equal(cdf,0)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
cdf = np.ma.filled(cdf_m,0).astype('uint8')
image3cdf = cdf[image3]
cv2_imshow(image3cdf)
plt.hist(image3cdf.ravel(),256,[0,256]); plt.show()












hist,bins = np.histogram(image3.flatten(),256,[0,256])
bin_centers = 0.5*(bins[1:] + bins[:-1])
from scipy import stats
pdf = stats.norm.pdf(bin_centers)

from matplotlib import pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(bin_centers, hist, label="Histogram of samples")
plt.plot(bin_centers, pdf, label="PDF")
plt.legend()
plt.show()









######################




```
